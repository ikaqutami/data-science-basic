# NumPy 
NumPy is a Linear Algebra Library for Python, it is so important for Data Science with Python because almost all of the libraries in the PyData Ecosystem rely on NumPy.

Numpy is also incredibly fast, as it has bindings to C libraries. For more info on why you would want to use Arrays instead of lists, check out this great [StackOverflow post](http://stackoverflow.com/questions/993984/why-numpy-instead-of-python-lists).

We will only learn the basics of NumPy.

### Numpy Array
Numpy arrays essentially come in two flavors: 
- vectors: 1-d arrays 
- matrices: 2-d arrays (but you should note a matrix can still have only one row or one column).

```python
## We can create an array by directly converting a list or list of lists:
my_list = [1,2,3]
my_list

np.array(my_list)
```

```python
my_matrix = [[1,2,3],[4,5,6],[7,8,9]]
my_matrix
```

```python
np.array(my_matrix)
```

### Built-in Methods
There are lots of built-in ways to generate Arrays:
- array: create arrays from raw Python sequences or use built-in initialization functions
- arange: return evenly spaced values within a given interval.
- zeros and ones: generate arrays of zeros or ones
- linspace: return evenly spaced numbers over a specified interval.
- eye: creates an identity matrix
- rand: create an array of the given shape and populate it with random samples from a uniform distribution over [0, 1).
- randn: return a sample (or samples) from the "standard normal" distribution.
- randint: return random integers from `low` (inclusive) to `high` (exclusive).
- reshape: returns an array containing the same data with a new shape.
- max,min,argmax,argmin: these are useful methods for finding max or min values. Or to find their index locations using argmin or argmax
- shape: an attribute that arrays have (not a method)
- dtype: check the data type of the object in the array


```python
# Create a 1D array (vector) from a list
arr_1d = np.array([1, 2, 3, 4])

# Create a 2D array (matrix)
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
```
